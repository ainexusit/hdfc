```python
6. Aggregation and Grouping
df.agg() – Apply aggregate functions to groups.
df.sum() – Sum of values for each column.
df.mean() – Mean of values for each column.
df.median() – Median of values for each column.
df.std() – Standard deviation of values.
df.min() – Minimum value for each column.
df.max() – Maximum value for each column.
df.count() – Number of non-NA/null entries.
```
